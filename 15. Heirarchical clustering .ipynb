There are 2 types of HC:
1. Agglomerative
2. Divise

**Agglomerative HC**:

Step 1: Make each data point as a single point cluster, that form N clusters.

Step 2: Take the two closest data points and make them one cluster, that form n-1 clusters.

Step 3: Take 2 closest clusters and make them one cluster. That forms N-2 clusters.

Step 4: Repeat step 3 until there is 1 cluster left.


**How to measure the distance between the 2 clusters?**
Distance between closest points, or furthest points, can be average of all the distances pairwaise, or the distance between the centroids.

**How to get to the right no. of clusters and where to stop?**
Creating the Dendograms(points vs euclidean distance). Dendogram is kind of memory of our HC  circle. The hieght determines the dissimilarity between the points/clusters. We can look at the horizontal levels and set the horizontal threshold, which indeed means setting threshold for the dissimilarity.
Or you can look for the Vertical largest distance.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('Mall_Customers.csv')
x = df.iloc[:, [3, 4]].values
```

We need to import scipy lib for plotting the dendogram! Ward is just like as we were minimising wcss, here we are minimising variance.

```python
#using dendogram to find optimum no. of class
import scipy.cluster.hierarchy as sch
dendo = sch.dendrogram(sch.linkage(x, method = 'ward'))

```

```python
#fitting the HC algo to the dataset.
from sklearn.cluster import AgglomerativeClustering
reg = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y = reg.fit_predict(x)
y
```

```python
plt.scatter(x[y==0, 0], x[y==0, 1], color = 'red', s =100)
plt.scatter(x[y==1, 0], x[y==1, 1], color = 'orange', s =100)
plt.scatter(x[y==2, 0], x[y==2, 1], color = 'green', s =100)
plt.scatter(x[y==3, 0], x[y==3, 1], color = 'magenta', s =100)
plt.scatter(x[y==4, 0], x[y==4, 1], color = 'blue', s =100)
```
